## Preprocessing

```python
from coppredict import preprocessing  as pr
import numpy as np
from coppredict.patricia import Patricia
from coppredict import evaluation as ev
from coppredict import prediction as pdt
from coppredict import util
np.random.seed(10) 
```

```python
path = '.'
filename = 'Results_ps_f_spmf_at_274132_bank_full_sizek_rt_50_.txt'
util.extract_patterns(filename)
#df_patterns = pr.load_patterns_file(filename, filename)
#df_patterns
```

```python
path = './weights'
filename = 'banks_weights2.csv'
weights = pr.load_weights(path, filename)
weights
```

```python
variable = 'coicop'
measure = 'ratio_sol'
df_patterns_weight = pr.calculate_weights_pattern(df_patterns, weights, variable, measure)
df_patterns_weight
```

```python
train, validate = pdt.splitTrainValidation(df_patterns_weight, 0.9)
```

```python
train = pr.orderBySublen(train)
train
```

## Patricia

```python
time_start = ev.getTime()
mem_start = ev.getProcessMemory()

X_trie = Patricia()

for i in range(len(train)):
    X_trie.addPattern(train.loc[i, "patterns"], round(train.loc[i, "weight"],4))

time_end = ev.getTime()
mem_end = ev.getProcessMemory()
```

```python
ev.printTimeBuild(time_start, time_end, "Patricia built in ")
```

```python
ev.printMemoryBuild(mem_start, mem_end, "Patricia built using ")
```

### Prediction

```python
validate = pdt.prepareValidationData(validate)
validate
```

```python
time_start = ev.getTime()
mem_start = ev.getProcessMemory()

cases, confidences, coverage = pdt.prediction(1, 10, 0.05, validate, X_trie)

time_end = ev.getTime()
mem_end = ev.getProcessMemory()
```
